#### Build df of words

```python
words = ''.join(pd_json[0].to_list()).split(' ')
words = [word.lower() for word in words ]

len(words)
```

```python
df_words = pd.DataFrame({'words':words})

vc_words = pd.DataFrame((df_words['words'].value_counts()))
```

#### Stop words

```python
import nltk
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
```

```python
len(stop_words)
```

```python
stop_words
```

#### Remove stop words

```python
import numpy as np
from IPython.display import display
```

```python
cond1 = (vc_words['words'] > 5)
cond2 = np.array([not(e) for e in vc_words.index.isin(stop_words)])
cond = cond1 * cond2
top_words = vc_words[cond]
```

```python
with pd.option_context('display.max_rows', 500):
    display(top_words.head(500))
```
